## What is Machine Learning?
Machine learning is a subset of artificial intelligence (AI) that enables computers to learn from experience (data) and make predictions or decisions without being explicitly programmed. Instead of relying on fixed rules, machine learning models identify patterns and relationships in data to improve their performance over time.

### How Machine Learning Works:
1. **Data Collection** – The model is trained on historical data, which can be structured (e.g., tables, spreadsheets) or unstructured (e.g., images, videos, text).
2. **Feature Extraction** – Important characteristics (features) of the data are identified to help the model make better predictions.
3. **Model Training** – The algorithm learns from the data by adjusting internal parameters to minimize errors.
4. **Testing & Validation** – The model is evaluated using unseen data to check its accuracy and reliability.
5. **Predictions & Continuous Learning** – The trained model makes predictions on new data and improves over time as more data is fed into the system.

### Key Characteristics of Machine Learning:
- **Automated Learning:** No need for explicit programming; the model learns patterns from data.
- **Adaptability:** It improves with more data and experience.
- **Pattern Recognition:** Finds hidden relationships within datasets.
- **Data-Driven Decision Making:** Uses statistical methods to make accurate predictions.

## Types of Machine Learning
### 1. Supervised Learning
**Definition:** In supervised learning, models learn from labeled datasets, meaning each input has a corresponding correct output. 
The model maps input features to the correct output by learning from past examples.

#### Applications of Supervised Learning:
- **Spam detection** – Classifying emails as spam or not spam.
- **Medical diagnosis** – Identifying diseases from medical images or records.
- **Stock price prediction** – Forecasting future prices based on historical data.
- **Speech recognition** – Converting spoken language into text.

### 2. Unsupervised Learning
**Definition:** In unsupervised learning, models work with unlabeled data. The goal is to find patterns, relationships, or structures within 
the dataset without predefined outputs.

#### Applications of Unsupervised Learning:
- **Customer segmentation** – Grouping customers based on behavior.
- **Anomaly detection** – Identifying fraudulent transactions in banking.
- **Data compression** – Reducing data dimensions while preserving key information.
- **Market basket analysis** – Understanding customer purchasing patterns.

### 3. Reinforcement Learning
**Definition:** Reinforcement learning is based on an agent interacting with an environment to learn an optimal strategy through trial and error. 
The agent receives rewards or penalties based on its actions.

#### Applications of Reinforcement Learning:
- **Autonomous vehicles** – Learning to navigate roads safely.
- **Robotics** – Training robots to perform tasks like assembling products.
- **Gaming** – AI mastering games like chess and Go.
- **Finance** – Algorithmic trading for maximizing returns.

## Importance of Statistics in Machine Learning
Statistics is fundamental in machine learning as it helps in:
- Understanding data distributions and variability.
- Identifying correlations and relationships between variables.
- Making data-driven decisions using probability theory.
- Evaluating model performance using statistical metrics (e.g., RMSE, accuracy, F1-score).
- Improving model predictions using statistical techniques.
